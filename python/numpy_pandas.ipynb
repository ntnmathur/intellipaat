```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [10, 20, 30]
})
print(df)
print("=============")
# Sum each column
col_sums = df.apply(sum)
print(col_sums)
```

```python
# Find the max value in each row
print(df)
print("=============")
row_max = df.apply(max, axis=1)
print(row_max)
```

```python
def custom_func(x):
    return x * 2 + 1

print(df)
print("=============")

df['C'] = df['A'].apply(custom_func)
print(df)
```

```python
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x + x, nums))
print(squared)  # Output: [1, 4, 9, 16]
```

```python
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)  # Output: [2, 4]
```
