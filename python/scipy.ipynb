```python
import numpy as np
c = np.array([15,18,20,26,32,38,32,24,21,16,13,11,14])
from scipy import stats
print(stats.zscore(c))

# z = x - u/sigma. x = value, u = mean, sigma = standard deviation
# zscore is between -3 and 3. Anything outside this is outlier
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0.0,1.0,0.1)

def f(x):
    return -np.exp(-(x-0.7)**2)
plt.plot(x,f(x),'o-')
```
