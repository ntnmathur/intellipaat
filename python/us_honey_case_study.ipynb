```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {'Cluster': ['Cluster 1', 'Cluster 2', 'Cluster 3'],
        'Metric 1': [10, 15, 20],
        'Metric 2': [25, 20, 15],
        'Metric 3': [30, 35, 40]}

df = pd.DataFrame(data)

# Melt the dataframe for plotting
df_melt = df.melt(id_vars='Cluster', var_name='Metric', value_name='Value')

# Create the snake plot
plt.figure(figsize=(8, 6))
for cluster in df['Cluster']:
    plt.plot(df_melt[df_melt['Cluster'] == cluster]['Metric'],
             df_melt[df_melt['Cluster'] == cluster]['Value'],
             marker='o')

plt.title('Snake Plot')
plt.xlabel('Metric')
plt.ylabel('Value')
plt.legend(df['Cluster'])
plt.show()
```
